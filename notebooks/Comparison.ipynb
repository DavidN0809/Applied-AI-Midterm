```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Define the directory containing the images
directory = "/opt/notebooks/Applied-AI-Midterm/figs"

# Create a figure and axes for the subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 10))

# Initialize counters for each row
count_a = 0
count_b = 0

# Iterate through all files in the directory, sorted alphabetically
for filename in sorted(os.listdir(directory)):
    # Check if the file is a .png file
    if filename.endswith(".png"):
        # Open the image file using PIL
        image = Image.open(os.path.join(directory, filename))
        # Check if the file contains "model-a" in the filename
        if "model-a" in filename and count_a < 3:
            # Plot the image in the top row
            axs[0][count_a].imshow(image)
            axs[0][count_a].axis('off')  # Remove axis ticks
            axs[0][count_a].set_title(filename)
            count_a += 1
        # Check if the file contains "model-b" in the filename
        elif "model-b" in filename and count_b < 3:
            # Plot the image in the bottom row
            axs[1][count_b].imshow(image)
            axs[1][count_b].axis('off')  # Remove axis ticks
            axs[1][count_b].set_title(filename)
            count_b += 1
        # Stop the loop after displaying 6 images
        if count_a == 3 and count_b == 3:
            break

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
#as you can see the Loss for model B is slighlty higher than model A
#as you can see the Accuracy for model B is slighlty lower than model A
#and the F1 score and AUC is slightly lower as well
#all of this indicated that Model A is slighlty better
```
